```python
import matplotlib.pyplot as plt
```

```python
dias = ['SEG', 'TER', 'QUA', 'QUI', 'SEX', 'SAB', 'DOM']
temperaturas = [24, 26, 22, 27, 31, 30, 19]

plt.figure(figsize=(6,4))
plt.plot(dias, temperaturas, marker='o', linestyle=':', color='red')
plt.title("Evolução da temperatura ao longo da semana")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
```

```python
meses = ['ABR', 'MAI', 'JUN', 'JUL', 'AGO', 'SET']
collor = [5, 14, 25, 20, 19, 22]
itamar = [10, 15, 21, 16, 14, 19]
fhc = [20, 22, 25, 28, 32, 36]

plt.figure(figsize=(10,6))
plt.plot(meses, collor, marker='o', linestyle=':', color='green', label='Fernando Collor de Melo')
plt.plot(meses, itamar, marker='x', linestyle=':', color='blue', label='Itamar Franco')
plt.plot(meses, fhc, marker='^', linestyle=':', color='red', label='Fernando Henrique Cardoso')

plt.title("Intenções de voto para Presidente da República")
plt.xlabel("Mês")
plt.ylabel("Intenção de votos (%)")

plt.legend()
plt.grid(True, linestyle='-', alpha=0.3)
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
vendas = [120, 95, 210, 80]

plt.figure(figsize=(8,4))

plt.bar(categorias, vendas, color=['red', 'green', 'orange', 'blue'])
plt.title("Vendas por categoria de produtos")
plt.xlabel("Categoria de produtos")
plt.ylabel("Quantidade vendida")

plt.ylim(0, 240)
plt.grid(True, linestyle='-', alpha=0.3)
plt.show()
```

```python
import numpy as np

np.random.seed(42)
notas = np.random.normal(70, 10, 500)

plt.figure(figsize=(8,5))
plt.hist(notas, bins=50, edgecolor='black', color='purple')

plt.title("Distribuição de notas da turma")
plt.xlabel('Notas')
plt.ylabel('Número de alunos')

plt.axvline(notas.mean(), color='red', linestyle='--', linewidth=2)
plt.legend(['Média das notas'])

plt.grid(True, linestyle=':', alpha=0.3)
plt.ylim(0,40)
plt.show()
```

```python
ativos = ['Ações', 'Renda fixa', 'Ouro', 'Criptomoedas']
valores = [45, 30, 15, 10]

plt.figure(figsize=(8,5))
explode = (0.1, 0, 0, 0)
plt.pie(valores, labels=ativos, startangle=90, autopct='%1.1f%%', explode=explode)
plt.title("Composição da carteira de investimentos")
plt.axis('equal')
plt.show()
```

```python
np.random.seed(0)
alturas = np.random.normal(170, 10, 50)
pesos = (alturas * 0.5) + np.random.normal(0, 5, 50)

plt.figure(figsize=(8,5))
plt.scatter(alturas, pesos, color='red', s=20)
plt.title("Relação entre altura e peso")
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True, linestyle=':', alpha=0.3)
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
df.shape
```

```python
sobreviventesPorGenero = df.groupby(['Sex', 'Survived']).size().unstack()
sobreviventesPorGenero.columns = ['Não Sobreviventes', 'Sobreviventes']
sobreviventesPorGenero.index = ['Mulheres', 'Homens']

fig, ax = plt.subplots(figsize=(10, 6))
sobreviventesPorGenero.plot(kind='bar', ax=ax)

plt.title("Sobreviventes e não sobreviventes do Titanic por gênero")
plt.xlabel('Gênero')
plt.ylabel('Número de pessoas')
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)

turma_1 = np.random.normal(75, 8, 30)
turma_2 = np.random.normal(68, 12, 30)
turma_3 = np.random.normal(82, 5, 30)

dados = [turma_1, turma_2, turma_3]

dic = { 'color': 'red', 'linewidth': 1, 'linestyle': '-' }

plt.figure(figsize=(8, 6))
plt.boxplot(dados, labels=['Turma 1', 'Turma 2', 'Turma 3'], patch_artist=True, medianprops=dic)
plt.title('Comparação da distribuição de notas entre turmas')
plt.ylabel('Nota')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
