```python
children = df.loc[df['Age'] < 18]
children
```

```python
firstClass = df.loc[df['Pclass'] == 1]
firstClass
```

```python
richChildren = df.loc[(df['Pclass'] == 1) & (df['Age'] < 18)]
richChildren
```

```python
survivor = df['Survived'].value_counts()
survivor
```

```python
survivorsBySex = df.groupby('Sex')['Survived'].mean() * 100
survivorsBySex
```

```python
meanSurvivedAge = df.groupby('Survived')['Age'].mean()
meanSurvivedAge
```

```python
dumbest = df.loc[df['Fare'] == df['Fare'].max()]
dumbest
```

```python
meanFareByClass = df.groupby('Pclass')['Fare'].mean()
meanFareByClass
```

```python
import matplotlib.pyplot as plt

survivorByClass = df.groupby('Pclass')['Survived'].sum()

plt.bar(survivorByClass.index, survivorByClass.values)
plt.xlabel("Classes")
plt.ylabel("Survivors")
plt.xticks(survivorByClass.index, [f"Class {int(c)}" for c in survivorByClass.index])
plt.title("Titanic survivor by class")
plt.show()

meanSurvivorByClass = df.groupby('Pclass')['Survived'].mean() * 100

plt.bar(meanSurvivorByClass.index, meanSurvivorByClass.values)
plt.xlabel("Classes")
plt.ylabel("Survivors")
plt.xticks(survivorByClass.index, [f"Class {int(c)}" for c in survivorByClass.index])
plt.title("Titanic survivor by class")
plt.show()
```
